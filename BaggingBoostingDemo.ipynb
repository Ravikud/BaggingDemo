```python
import pandas as pd
import numpy as np
```

```python
Df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
Df.head(10)
```

```python
Df.info()
```

```python
Df.species.value_counts()
```

```python
#Balanced Dataset each label contsins equal records
features = Df.iloc[:,:-1].values
label = Df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size=0.2,
                                                random_state=10)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
#Bagging
model1 = LogisticRegression()
from sklearn.ensemble import BaggingClassifier
bagClassifier = BaggingClassifier(base_estimator=model1,#Algo to predict(in this case. Log Reg
                                n_estimators=101)
```

```python
bagClassifier.fit(X_train,y_train)
```

```python
print(bagClassifier.score(X_train,y_train))
print(bagClassifier.score(X_test,y_test))
```

```python
#Underfitting model not many changes when bagging is used
from mlxtend.plotting import plot_decision_regions
```

```python
dataLabels = pd.DataFrame(y_train).replace(['setosa','versicolor','virginica'],[1,2,3])
```

```python
finalLabels = dataLabels.values
```

```python
finalLabels
```

```python
# Plotting decision regions
plot_decision_regions(X_train, finalLabels, clf=model, legend=2)
```

```python
from sklearn.ensemble import AdaBoostClassifier
modelAda = LogisticRegression()
boostAlgo = AdaBoostClassifier(base_estimator=modelAda,#Algo to predict(in this case. Log Reg
                                n_estimators=101)
```

```python
boostAlgo.fit(X_train,y_train)
```

```python
print(boostAlgo.score(X_train,y_train))
print(boostAlgo.score(X_test,y_test))
```
